# Visualizing statitical relationships

One useful function of seaborn package is to visualize relationships between variables - `replot()`. It also generate two sub-functions:

- `scatterplot()` (with `kind = "scatter"`)
- `lineplot()` (with `kind = "line"`)



```
# load a sample data
tips = sns.load_dataset('tips')
sns.relplot(x = 'total_bill', y = 'tip',hue = 'smoker', data = tips)
```

A number things we notice from the above plot. It is similar to the Altair encode part, but here seaborn puts all arguments together into one function. Note 'color' is controled by 'hue'. 

You can also change the mark shapes using 'style':

```
sns.relplot(x = 'total_bill', y = 'tip', hue = 'smoker', style = 'smoker', data = tips);
```

```
# an example of random walk graph
df = pd.DataFrame(dict(time=np.arange(500),
                       value=np.random.randn(500).cumsum()))
g = sns.relplot(x="time", y="value", kind="line", data=df)
```

## Aggregation and representing uncertainty

More complex datasets will have multiple measurements for the same value of the x variable. The default behavior in seaborn is to aggregate the multiple measurements at each x value by plotting the mean and the 95% confidence interval around the mean:

```
fmri = sns.load_dataset("fmri")
sns.relplot(x="timepoint", y="signal", kind="line", data=fmri);
```

The above plot shows the 95% confidence interval based on the raw data. 


```
# glimpse the head of the table
fmri.head()
```

```
# show the summary of the table
fmri.describe(include = ['object'])
```

If you want to plot the mean only, you need `ci = None`.


```
sns.relplot(x="timepoint", y="signal", ci=None, kind="line", data=fmri);
```

From the summary output we know the data have two events. We can visualize them in one figure using two different colors:

```
sns.relplot(x="timepoint", y="signal", hue="event", kind="line", data=fmri);
```

# Bar plots and Point plots

There is a simplified function called `barplot()`. Basically, it is the same when you use `kind = "bar"` in `catplot()`.

# Visualizing linear relationship

Two main functions in seaborn are used to visualize a linear relationship as determined through regression. These functions, `regplot()` and `lmplot()` are closely related, and share much of their core functionality. It is important to understand the ways they differ, however, so that you can quickly choose the correct tool for particular job.

```
#sns.regplot(x="total_bill", y="tip", data=tips);
sns.lmplot(x="total_bill", y="tip", data=tips);
```

`lmplot()` has data as a required parameter and the x and y variables must be specified as strings. It has some additional features. Please check their differences in the official website. 

```
sns.lmplot(x="size", y="tip", data=tips, x_estimator=np.mean);
```

You can fit other models other than simple linear model (e.g., logistic regression)

```
tips["big_tip"] = (tips.tip / tips.total_bill) > .15
sns.lmplot(x="total_bill", y="big_tip", data=tips,
           logistic=True, y_jitter=.03);
```

# Conditioning on other variables

 Often, however, a more interesting question is “how does the relationship between these two variables change as a function of a third variable?” This is where the difference between regplot() and lmplot() appears. While regplot() always shows a single relationship, lmplot() combines regplot() with FacetGrid to provide an easy interface to show a linear regression on “faceted” plots that allow you to explore interactions with up to three additional categorical variables.

```
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips);
```
